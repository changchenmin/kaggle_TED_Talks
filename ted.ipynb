```python
# 處理時間: 秒數 -> 格林威治時區
# 步驟一: 先處理一格
from datetime import datetime
# datetime.utcfromtimestamp(1140825600)
str(datetime.utcfromtimestamp(1140825600))
```

```python
# 步驟二: 再將一個的內容依序走過其他格子
# 處理思維: 定義轉換流程 -> 遞給整行 -> 每一個走過流程
def timetrans(s):
    return str(datetime.utcfromtimestamp(s))
# apply: apply(流程名稱) -> 對每一個執行一次流程
df["filmutc"] = df["film_date"].apply(timetrans)
df
```

```python
# 篩選操作: 原理 df[[跟你的列一樣多的布林]]
df.head(3)[[True, False, True]]
```

```python
# 篩選: 轉換操作 -> 2550 True False
# comments: 3000以上 & 5000以下
def fflow(n):
    if n > 3000 and n < 5000:
        return True
    else:
        return False
f = df["comments"].apply(fflow)
df[f]
```

```python
# eval() -> 把傳入的字串，當成一個python程式執行

a = eval("[1, 2, 3]")
# a = [1, 2, 3]
a
```

```python
# 搜尋tags裡面的關鍵詞
# 轉換操作 -> 篩選
def fflow(s):
    taglist = eval(s)
    if "computers" in taglist:
        return True
    else:
        return False
df[df["tags"].apply(fflow)]
```
